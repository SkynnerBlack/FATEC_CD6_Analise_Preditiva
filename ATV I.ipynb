### 3. Use outra distribuição a priori no exemplo anterior e faça os gráficos usando o matplotlib e o arviz.

```python
n = len(moedas)  # Número total de observações
k = sum(moedas)  # Número de sucessos (exemplos de 1s)

with pm.Model() as model:
    # Distribuição a priori Beta para theta
    alpha = 2
    beta = 2
    theta = pm.Beta("theta", alpha, beta)

    # Distribuição Binomial para a variável observada
    p = pm.Binomial("p", n=n, p=theta, observed=k)

    # Amostragem
    trace = pm.sample(1000, return_inferencedata=False)
```

```python
# Traceplot de theta
plt.figure(figsize=(10, 5))
plt.plot(trace['theta'])
plt.title('Traceplot of theta')
plt.xlabel('Sample')
plt.ylabel('Theta')
plt.show()
```

```python
# Resumo do trace
summary = az.summary(trace)
print(summary)

# Plotando a distribuição posterior de theta
az.plot_posterior(trace, var_names=['theta'])
plt.show()
```
